# ポリフェノール類のデータの呼び出しとデータを確認

```python
!wget https://raw.githubusercontent.com/chemoinformatics-lecture/lecture-beginner/main/lesson02_jupyter/data/compounds-structures.csv
```

```python
import pandas as pd
df = pd.read_csv('/content/compounds-structures.csv', sep=',')
df
```

# tsne

```python
from chemplot import Plotter
```

```python
from chemplot import Plotter

cp_tsne=Plotter.from_smiles(df["smiles"],	target = df["compound_class"], target_type="C")

df_tsne = cp_tsne.tsne()

cp_tsne.visualize_plot()
```

```python
df_tsne
```

# PCA

```python
cp_pca=Plotter.from_smiles(df["smiles"],	target = df["compound_class"], target_type="C")
df_pca = cp_pca.pca()
cp_pca.visualize_plot(size=8)
```

```python
df_pca
```

# UMAP

## tailored (mordredの記述子200個くらいで行ったもの)

```python
cp_umap=Plotter.from_smiles(df["smiles"], target = df["compound_class"], target_type="C", sim_type="tailored")

df_umap = cp_umap.umap(n_neighbors=5,min_dist=0.99,pca=True)

cp_umap.visualize_plot()
```

### structural　(ECFP(morgan fingerprint)のradius 2 bit 2048で行ったもの)

```python
cp_umap=Plotter.from_smiles(df["smiles"], target = df["compound_class"], target_type="C", sim_type="structural")

df_umap = cp_umap.umap(n_neighbors=5,min_dist=0.99,pca=True)

cp_umap.visualize_plot()
```
